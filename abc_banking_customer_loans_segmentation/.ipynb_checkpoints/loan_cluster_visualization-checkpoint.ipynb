# Loan K-Means Cluster Visualization

This notebook visualizes the clustered output exported from **Assign Data to Clusters** in Azure ML Designer.

It uses the **`Assignments`** column as the cluster label and builds visuals that help explain what each customer segment looks like.


```python

from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.grid"] = True

# Try common locations
candidate_paths = [
    Path("../datasets/loan_data.csv"),
    Path("/mnt/data/data.csv"),
]

csv_path = None
for p in candidate_paths:
    if p.exists():
        csv_path = p
        break

if csv_path is None:
    raise FileNotFoundError("Could not find data.csv. Put the exported clustered CSV beside this notebook or in /mnt/data/data.csv")

df = pd.read_csv(csv_path)
print(f"Loaded: {csv_path}")
print(f"Shape: {df.shape}")
df.head()
```

```python

# Detect key columns
cluster_col = "Assignments"
if cluster_col not in df.columns:
    raise ValueError("Expected an 'Assignments' column in the clustered export.")

distance_cols = [c for c in df.columns if c.startswith("DistancesToClusterCenter")]
one_hot_cols = [c for c in df.columns if "-" in c and c not in distance_cols]

categorical_cols = [
    c for c in ["EmploymentStatus", "EducationLevel", "MaritalStatus", "HomeOwnershipStatus", "LoanPurpose"]
    if c in df.columns
]

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
base_numeric_cols = [c for c in numeric_cols if c not in one_hot_cols + distance_cols + [cluster_col]]

print("Cluster column:", cluster_col)
print("Distance columns:", distance_cols)
print("Categorical columns:", categorical_cols)
print("Base numeric columns:", base_numeric_cols[:20], "..." if len(base_numeric_cols) > 20 else "")
```

## 1) Cluster sizes

This shows how many customers ended up in each segment.


```python

cluster_counts = df[cluster_col].value_counts().sort_index()

ax = cluster_counts.plot(kind="bar")
ax.set_title("Customer Count by Cluster")
ax.set_xlabel("Cluster")
ax.set_ylabel("Number of Customers")

for i, v in enumerate(cluster_counts.values):
    ax.text(i, v + max(cluster_counts.values)*0.01, f"{v}", ha="center")

plt.tight_layout()
plt.show()

cluster_counts
```

## 2) Cluster compactness from Azure ML output

Lower distance to own center generally means the cluster is tighter / more compact.


```python

if distance_cols:
    # For each row, the "own center" distance is the distance column matching its cluster id.
    own_distance = df.apply(lambda r: r[f"DistancesToClusterCenter no.{int(r[cluster_col])}"], axis=1)
    compactness = own_distance.groupby(df[cluster_col]).agg(["mean", "median", "max"]).rename(
        columns={"mean": "avg_distance_to_own_center", "median": "median_distance", "max": "max_distance"}
    )
    compactness
else:
    print("No distance columns found.")
```

## 3) Business profile of each cluster

The export contains many transformed and one-hot-encoded columns.  
This section focuses on the original business fields that are easiest to interpret.


```python

preferred_numeric = [
    "Age", "AnnualIncome", "CreditScore", "Experience", "LoanAmount", "LoanDuration",
    "MonthlyDebtPayments", "CreditCardUtilizationRate", "NumberOfOpenCreditLines",
    "NumberOfCreditInquiries", "DebtToIncomeRatio", "PaymentHistory",
    "LengthOfCreditHistory", "SavingsAccountBalance", "CheckingAccountBalance",
    "TotalAssets", "TotalLiabilities", "MonthlyIncome", "UtilityBillsPaymentHistory",
    "JobTenure", "NetWorth", "BaseInterestRate", "InterestRate",
    "MonthlyLoanPayment", "TotalDebtToIncomeRatio", "RiskScore"
]
profile_numeric = [c for c in preferred_numeric if c in base_numeric_cols]

cluster_profile = df.groupby(cluster_col)[profile_numeric].mean().round(3)
cluster_profile
```

```python

# Dominant category per cluster
dominant_categories = {}
for c in categorical_cols:
    dominant_categories[c] = df.groupby(cluster_col)[c].agg(lambda s: s.mode().iat[0] if not s.mode().empty else np.nan)

dominant_categories_df = pd.DataFrame(dominant_categories)
dominant_categories_df
```

## 4) Heatmap of cluster means

This is one of the best ways to see how the clusters differ across numeric features.

**Interpretation**
- warmer colors = relatively higher average within that cluster
- cooler colors = relatively lower average within that cluster


```python

# Standardize cluster means only for display so features with larger scales do not dominate the heatmap
heatmap_df = cluster_profile.copy()
heatmap_scaled = (heatmap_df - heatmap_df.mean()) / heatmap_df.std(ddof=0)

fig, ax = plt.subplots(figsize=(14, 7))
im = ax.imshow(heatmap_scaled.values, aspect="auto")

ax.set_xticks(range(len(heatmap_scaled.columns)))
ax.set_xticklabels(heatmap_scaled.columns, rotation=75, ha="right")
ax.set_yticks(range(len(heatmap_scaled.index)))
ax.set_yticklabels([f"Cluster {i}" for i in heatmap_scaled.index])

ax.set_title("Relative Cluster Profile Heatmap")
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Relative Mean (z-score across clusters)")

plt.tight_layout()
plt.show()
```

## 5) Box plots for key business features by cluster

These help show the spread and not just the average.


```python

key_features = [c for c in ["RiskScore", "CreditScore", "AnnualIncome", "LoanAmount", "DebtToIncomeRatio", "MonthlyIncome", "NetWorth"] if c in df.columns]

for feature in key_features:
    fig, ax = plt.subplots(figsize=(8, 5))
    data = [df.loc[df[cluster_col] == k, feature].dropna().values for k in sorted(df[cluster_col].unique())]
    ax.boxplot(data, tick_labels=[f"Cluster {k}" for k in sorted(df[cluster_col].unique())], showfliers=False)
    ax.set_title(f"{feature} by Cluster")
    ax.set_xlabel("Cluster")
    ax.set_ylabel(feature)
    plt.tight_layout()
    plt.show()
```

## 6) PCA scatter plot of customers

K-means works in high-dimensional space, which is hard to visualize directly.  
PCA compresses the numeric features into 2 dimensions so you can see approximate cluster separation.


```python

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca_features = [c for c in base_numeric_cols if c not in ["LoanApproved"]]
X = df[pca_features].copy().fillna(df[pca_features].median(numeric_only=True))

# Standardize again for PCA display
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plot_df = pd.DataFrame({
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1],
    "Cluster": df[cluster_col].values
})

fig, ax = plt.subplots(figsize=(9, 6))
for cluster_id in sorted(plot_df["Cluster"].unique()):
    part = plot_df[plot_df["Cluster"] == cluster_id]
    ax.scatter(part["PC1"], part["PC2"], s=15, alpha=0.5, label=f"Cluster {cluster_id}")

ax.set_title("PCA View of Customer Segments")
ax.set_xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)")
ax.set_ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)")
ax.legend()
plt.tight_layout()
plt.show()
```

## 7) Categorical mix inside each cluster

These charts show whether a cluster is dominated by certain categories such as loan purpose, home ownership, or education level.


```python

for cat_col in categorical_cols:
    ctab = pd.crosstab(df[cluster_col], df[cat_col], normalize="index")
    ax = ctab.plot(kind="bar", stacked=True, figsize=(10, 6))
    ax.set_title(f"{cat_col} Distribution by Cluster")
    ax.set_xlabel("Cluster")
    ax.set_ylabel("Proportion")
    plt.legend(title=cat_col, bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    plt.show()
```

## 8) Quick interpretation template

Use the tables above to label the clusters in business terms. A practical way to write it is:

- **Cluster 0:** describe its income, credit score, debt burden, and risk score pattern
- **Cluster 1:** describe its dominant customer traits and likely financial profile
- **Cluster 2:** describe whether it looks more stable, mainstream, or higher-risk

Do not label a cluster as “good” or “bad” just from the number.  
Label it from the actual profile you see in the charts and summary tables.
