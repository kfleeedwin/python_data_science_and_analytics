<h4>1. Impute missing values in the dataset columns.</h4>

```python
# list of columns to be kept (15)
cols_to_keep = [
    'iso_code', 'continent', 'location',
    'date', 'total_cases', 'new_cases',
    'total_deaths', 'new_deaths', 'total_tests',
    'new_tests', 'positive_rate', 'male_smokers',
    'female_smokers','people_vaccinated',
    'hospital_beds_per_thousand'
]

owid_df = owid_df[cols_to_keep]  # filter the DataFrame to keep only the specified columns
owid_df.info()                   # display information about the DataFrame
```

<h4>Missing Value Imputation Strategy</h4>

- Identifier / categorical columns (do not impute): iso_code, continent, location

- Time-series cumulative metrics (forward-fill per location): total_cases, total_deaths, total_tests, people_vaccinated

- Daily flow metrics (forward-fill, then fill remaining with 0): new_cases, new_deaths, new_tests

- Ratio / rate metrics (forward-fill): positive_rate

- Static country-level indicators (forward + backward fill per location): male_smokers, female_smokers, hospital_beds_per_thousand


```python
# Impute cumulative metrics
owid_df = owid_df.sort_values(['location', 'date'])  # sort by location and date to ensure chronological order within each country

cumulative_cols = [                                  # define columns that contain cumulative data
    'total_cases', 'total_deaths',
    'total_tests', 'people_vaccinated'
]

owid_df[cumulative_cols] = (                         # forward-fill missing values in cumulative columns within each location group
    owid_df.groupby('location')[cumulative_cols]     # this propagates the last known value forward until a new value is encountered
    .ffill()
)

print(owid_df[cumulative_cols].isnull().sum())
```

```python
# Impute daily flow metrics
flow_cols = ['new_cases', 'new_deaths', 'new_tests']  # Define columns representing daily flow metrics

owid_df[flow_cols] = (                                # forward-fill missing values within each location group
    owid_df.groupby('location')[flow_cols]            # this propagates the last known value forward until a new value is found
    .ffill()
)

owid_df[flow_cols] = owid_df[flow_cols].fillna(0)     # replace any remaining NaN values with 0

print(owid_df[flow_cols].isnull().sum())
```

```python
# Impute rate metric
owid_df['positive_rate'] = (                          # forward-fill missing values in 'positive_rate' column for each location group
    owid_df.groupby('location')['positive_rate']      # this propagates the last valid observation forward to next valid observation
    .ffill()
)

print(owid_df['positive_rate'].isnull().sum())
```

```python
# Impute static country indicators
static_cols = [                                       # define columns with static country indicators that rarely change over time
    'male_smokers',
    'female_smokers',
    'hospital_beds_per_thousand'
]

""" Impute missing values in static columns by:
1. Grouping data by country ('location')
2. For each country, applying forward fill (ffill) to propagate last valid observation forward
3. Then applying backward fill (bfill) to propagate first valid observation backward
This assumes these indicators don't change frequently and missing values can be filled from nearby dates
"""
owid_df[static_cols] = (
    owid_df.groupby('location')[static_cols]
    .transform(lambda x: x.ffill().bfill())           # lambda is an anonymous function - here it takes input x (each column in each group)
)                                                     # and applies ffill() then bfill() to impute missing values

print(owid_df[static_cols].isnull().sum())
```

<h4>3. Create new features if needed (e.g., extract year and month from the 'date' column).</h4>

```python
owid_df = owid_df.sort_values(['location', 'date'])         # sort by location and date to ensure chronological order within each country

""" year → long-term trends
    month / month_name → seasonality
    year_month → clean monthly grouping without date noise """

owid_df['year'] = owid_df['date'].dt.year                   # extract year and month from the 'date' column
owid_df['month'] = owid_df['date'].dt.month
owid_df['month_name'] = owid_df['date'].dt.month_name()     # extract month name
owid_df['year_month'] = owid_df['date'].dt.to_period('M')   # combined year–month, useful for aggregation
```

```python
# Calculate 7-day rolling average for new COVID-19 cases
# 1. Group data by location to ensure averages are calculated separately for each country/region
# 2. Apply rolling window of 7 days (with minimum 1 observation required) to smooth daily fluctuations
# 3. Use transform to maintain the original dataframe structure

owid_df['new_cases_7d_avg'] = (
    owid_df.groupby('location')['new_cases']
    .transform(lambda x: x.rolling(window=7, min_periods=1).mean())
)

# 7-day rolling average for new deaths
# Using the same approach as above but for death data
# This helps smooth out reporting inconsistencies and weekend effects
owid_df['new_deaths_7d_avg'] = (
    owid_df.groupby('location')['new_deaths']
    .transform(lambda x: x.rolling(window=7, min_periods=1).mean())
)
```

```python
# Monthly aggregation
# Calculate the sum of new COVID cases for each location and month
# - Groups the data by location and year_month columns
# - Applies the sum function to the new_cases column within each group
# - Uses transform() to return results with the same shape as the original dataframe
# - This creates a new column where each row contains the monthly total for its location

owid_df['monthly_new_cases'] = (
    owid_df.groupby(['location', 'year_month'])['new_cases']
    .transform('sum')
)
```

<h4>2. Visualize the distribution of total COVID-19 cases by year using a box plot.</h4>

```python
box_df = owid_df[owid_df['continent'].notna()].copy()     # keep only real countries

yearly_country_cases = (                                  # get yearly total cases per country
    box_df
    .sort_values('date')
    .groupby(['location', 'year'])['total_cases']
    .last()
    .reset_index()
)

plt.figure(figsize=(12, 6))
plt.yscale('log')                                          # set y-axis to logarithmic scale to better visualize wide range of case numbers

ax = sns.boxplot(
    data=yearly_country_cases,
    x='year',
    y='total_cases',
    showfliers=False,
    notch=True,
    width=0.6
)

medians = (                                                 # calculate median per year
    yearly_country_cases
    .groupby('year')['total_cases']
    .median()
)

for i, year in enumerate(medians.index):                     # annotate median values in full digits
    median_value = medians.loc[year]
    ax.text(
        i,
        median_value,
        f"{int(median_value):,}",
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# set title and labels
plt.title('Distribution of Total COVID-19 Cases by Year', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Total Cases (log scale)')
plt.xticks(rotation=45)
formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)
plt.tight_layout()
plt.show()
```

```python
# Use real countries only
df = owid_df[owid_df['continent'].notna()].copy()

# Aggregate globally per day
global_daily = (
    df.groupby('date')[['new_tests', 'new_cases']]
    .sum()
    .reset_index()
    .sort_values('date')
)

# Build cumulative totals
global_daily['total_tests'] = global_daily['new_tests'].cumsum()
global_daily['total_cases'] = global_daily['new_cases'].cumsum()

# Remove zero rows
global_daily = global_daily[(global_daily['total_tests'] > 0) & (global_daily['total_cases'] > 0)]

# Positivity rate
global_daily['positivity_rate'] = global_daily['total_cases'] / global_daily['total_tests']
global_daily = global_daily[global_daily['positivity_rate'] <= 1]

# Add year for coloring
global_daily['year'] = pd.to_datetime(global_daily['date']).dt.year

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

ax = sns.scatterplot(
    data=global_daily,
    x='total_tests',
    y='positivity_rate',
    alpha=0.4,
    color='salmon',
    hue='year',
    palette='viridis',
    s=40
)

plt.xscale('log')
plt.xlabel('Total Tests Conducted (log scale)')
plt.ylabel('Positivity Rate (total cases / total tests)')
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y*100:.1f}%'))
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x/1e6:.0f}M'))
plt.title('Global Positivity Rate vs Total Tests Conducted', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Keep real countries only
df = owid_df[owid_df['continent'].notna()].copy()

# Latest record per country
latest = (
    df.sort_values('date')
      .groupby('location')
      .tail(1)
)

# Remove invalid rows
latest = latest[
    (latest['total_cases'] > 0) &
    (latest['total_deaths'] > 0)
].copy()

# Fatality rate
latest['fatality_rate'] = latest['total_deaths'] / latest['total_cases']

# Keep smoking + CFR
smoking_df = latest[['location','fatality_rate','male_smokers','female_smokers']].dropna()

sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))

ax = sns.regplot(
    data=smoking_df,
    x='male_smokers',
    y='fatality_rate',
    scatter_kws={'alpha':0.6},
    line_kws={'color':'red'},
    color='darkblue'
)

ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y*100:.2f}%'))

plt.title('Fatality Rate vs Male Smoking Prevalence (by Country)', fontsize=14)
plt.xlabel('Male Smokers (%)')
plt.ylabel('Fatality Rate')

interpretation = (
    "The scatter plot shows a weak association between male smoking prevalence and COVID-19 fatality rate across "
    "countries. The nearly flat regression line and wide dispersion of points indicate that smoking prevalence "
    "alone does not strongly predict national fatality rates. While smoking is a known individual-level risk "
    "factor for respiratory illness, cross-country COVID-19 outcomes appear to be more strongly influenced by "
    "other factors such as healthcare capacity, age distribution, testing coverage, vaccination rates, and reporting practices."
)

plt.figtext(
    0.5,          # x position (center)
    -0.15,        # y position (below plot)
    interpretation,
    wrap=True,
    horizontalalignment='center',
    fontsize=10
)

plt.tight_layout(rect=[0, 0.15, 1, 1])
plt.show()
```

```python
plt.figure(figsize=(10, 5))

ax = sns.regplot(
    data=smoking_df,
    x='female_smokers',
    y='fatality_rate',
    scatter_kws={'alpha':0.6},
    line_kws={'color':'red'},
    color='yellowgreen'
)

ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y*100:.2f}%'))

plt.title('Fatality Rate vs Female Smoking Prevalence (by Country)', fontsize=14)
plt.xlabel('Female Smokers (%)')
plt.ylabel('Fatality Rate')

interpretation = (
    "The scatter plot shows a weak inverse relationship between female smoking prevalence and COVID-19 fatality "
    "rate across countries. The slight downward trend suggests that countries with higher female smoking rates "
    "tend to exhibit marginally lower fatality rates; however, the wide dispersion of points indicates that "
    "smoking prevalence alone does not meaningfully explain differences in national COVID-19 outcomes. "
    "These results highlight that factors such as healthcare capacity, population age structure, testing "
    "coverage, vaccination rates, and reporting practices likely play a much larger role in determining "
    "fatality rates than smoking prevalence at the country level."
)

plt.figtext(
    0.5,          # x position (center)
    -0.15,        # y position (below plot)
    interpretation,
    wrap=True,
    horizontalalignment='center',
    fontsize=10
)

plt.tight_layout(rect=[0, 0.15, 1, 1])
plt.tight_layout()
plt.show()
```

<h4>4. Create a heatmap to analyze the relationship between hospital beds per thousand and fatality rate. </h4>

```python
beds_df = owid_df[owid_df['continent'].notna()].copy()

latest_country_df = (
    beds_df
    .sort_values('date')
    .groupby('location')
    .tail(1)
)

latest_country_df['fatality_rate'] = (
    latest_country_df['total_deaths'] /
    latest_country_df['total_cases']
)

latest_country_df = latest_country_df.replace(
    [float('inf'), -float('inf')], None
)

analysis_df = latest_country_df[
    ['hospital_beds_per_thousand', 'fatality_rate']
].dropna()

correlation_matrix = analysis_df.corr()
correlation_matrix

plt.figure(figsize=(8, 6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    square=True,
    linewidths=0.5
)

plt.title(
    'Correlation Between Hospital Beds per Thousand and COVID-19 Fatality Rate',
    fontsize=13
)

interpretation = (
    "The heatmap shows a weak negative correlation (−0.24) between hospital beds per thousand and COVID-19 "
    "fatality rate, indicating that countries with higher hospital bed capacity tend to experience slightly "
    "lower fatality rates. However, the modest magnitude of the correlation suggests that hospital bed "
    "availability alone does not strongly determine national COVID-19 outcomes, highlighting the influence "
    "of additional factors such as age structure, healthcare quality, testing coverage, vaccination rates, "
    "and reporting practices."
)

plt.figtext(
    0.5,          # x position (center)
    -0.15,        # y position (below plot)
    interpretation,
    wrap=True,
    horizontalalignment='center',
    fontsize=10
)

plt.tight_layout(rect=[0, 0.15, 1, 1])
plt.show()
```
