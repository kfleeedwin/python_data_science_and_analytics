<h3>Task 1: Load the Dataset and Explore</h3>

```python
import pandas as pd               # imports the pandas library with the alias 'pd', which is used for data manipulation and analysis.
import matplotlib.pyplot as plt   # imports the pyplot module from matplotlib with the alias 'plt', which is used for creating visualizations.
import seaborn as sns             # imports the seaborn library with the alias 'sns', which is a statistical data visualization library based on matplotlib.
import requests                   # imports the requests library, which is used to make HTTP requests to fetch data from URLs
from io import StringIO           # imports StringIO from io module, which helps in treating strings as file-like objects

# URL where the Titanic dataset CSV is hosted
url = "https://claaslms.educlaas.com/asset-v1:EDUCLaaS+PYD+PDDS-PYD-1125-10Jan2026B+type@asset+block/titanic.csv"

headers = {
    "User-Agent": "Mozilla/5.0"   # sets a user agent header to mimic a browser request, which can help avoid request blocking
}
response = requests.get(url, headers=headers)       # makes an HTTP GET request to the URL with the specified headers and stores the response

# titanic_df = pd.read_csv(url)                     # open the url link and reads a csv file named "titanic.csv" into a pandas dataframe object named 'titanic_df'.
titanic_df = pd.read_csv(StringIO(response.text))   # converts the text response to a file-like object and loads it into a pandas dataframe.
titanic_df.head()                                   # displays the first 5 rows of the dataframe
```

-----------------------------
<h3>Task 2: Visualize Passenger Survival</h3>

```python
sns.set_style('whitegrid')                                                        # set the visual style
plt.figure(figsize=(8, 5))                                                        # plot window, set width = 8, height = 5
ax = sns.countplot(x='Survived', data=titanic_df, edgecolor='black', hue='Sex')   # bar chart showing how many rows fall into each category of the Survived column, assign the plot to 'ax'
                                                                                  # counts occurrences of categorical values by Sex
for container in ax.containers:
    ax.bar_label(container)                                                       # add labels to each bar container

# add labels and title
plt.xlabel('Survival Status (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.title('Passenger Survival Counts')

# show the plot
plt.show()
```

Observations:
- There is a significant imbalance between those who survived and those who did not.
- A much higher number of passengers are recorded under the "No" (0) status than the "Yes" (1) status.
- This plot highlights a class imbalance.

<h3>Task 3: Analyze Passenger Age Distribution</h3>

```python
sns.set_style('whitegrid')                     # set the visual style
plt.figure(figsize=(8, 6))                     # plot window, set width = 8, height = 6

# use histogram function to visualize the distribution of ages
# drop null values with `.dropna()`
# set the number of bins to 15 for better visibility of the distribution
plt.hist(titanic_df['Age'].dropna(), bins=15, color='yellowgreen', edgecolor='black')

# add labels and title
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Ages on the Titanic')

# show the plot
plt.show()
```

Observations:
- Dominant Age Group: The largest group of passengers was between 20 and 30 years old.
- Presence of Children: There is a noticeable "spike" at the very beginning of the x-axis (ages 0–5), indicating a significant number of infants and young children on board.
- Right-Skewed Distribution: The data is "right-skewed," meaning that while there are many young people, the numbers drop off steadily as age increases. There were very few elderly passengers, with almost no one over the age of 70.

-----------------------------
<h3>Task 4: Explore Class-wise Distribution</h3>

```python
sns.set_style('whitegrid')                     # set the visual style
plt.figure(figsize=(6, 4))                     # plot window, set width = 6, height = 4

# use seaborn's countplot to visualize the frequency of passengers in each class
# apply hue to each class, grouped by color
# apply the 'plasma' color palette to make the bars visually distinct
sns.countplot(x='Pclass', data=titanic_df, hue='Pclass', palette='plasma')

# add labels and title
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Class')

# show the plot
plt.show()
```

Observations:
- Dominance of 3rd Class: The vast majority of passengers were in 3rd Class (represented by the yellow bar).
- Socio-economic Insight: The high volume of 3rd-class passengers reflects the large number of immigrants and lower-income families traveling on the ship.
- 1st Class vs. 2nd Class: There were more passengers in 1st Class than in 2nd Class. 2nd Class was the smallest group on the ship.

<h3>Task 5: Investigate Fare Distribution</h3>

```python
sns.set_style('whitegrid')                     # set the visual style
plt.figure(figsize=(8, 4))                     # plot window, set width = 8, height = 4

# use Seaborn's boxplot to visualize the distribution of passenger fares
# set the orientation to horizontal by using the 'x' parameter
sns.boxplot(x='Fare', data=titanic_df, color='skyblue')

# add labels and title
plt.xlabel('Fare (in pounds)')
plt.title('Distribution of Passenger Fares')

# show the plot
plt.show()
```

The resulting box plot show:
- Box → middle 50% of fares (IQR)
- Line inside box → median fare
- Whiskers → typical min/max (excluding outliers), the range of typical fares
- Dots → outliers (unusually high fares), (individual points beyond the whiskers)

Observations:
- Most passengers paid very low fares, The majority of passengers were third class
- A small number of passengers paid very high fares, classic income inequality pattern
- A few wealthy passengers paid dramatically more than everyone else, first-class luxury tickets

<h3>Task 6: Correlation Between Variables</h3>

```python
# select only numerical columns from the dataset using `select_dtypes`
# compute the correlation matrix of numerical variables
correlation = titanic_df.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(10, 8))       # plot window, set width = 10, height = 8
sns.heatmap(correlation,          # use seaborn's heatmap to visualize the correlation matrix
            annot=True,           # show correlation values
            cmap='coolwarm',      # color scheme
            linewidths=0.5,       # add lines between cells
            fmt='.2f')            # format numbers to 2 decimal places

plt.title('Correlation Heatmap of Titanic Dataset', fontsize=14)  # add title

plt.tight_layout()                # automatically adjust the spacing, to fit perfectly within the figure area
plt.show()                        # show the plot
```

Observations:
- Survival is most related to Pclass and Fare, 1st class passengers survived more, Higher fare, better survival odds.
- Family-related features are linked. People traveling with siblings often had parents/children too.

<h3>Task 7: Visualize Passenger Class and Age</h3>

```python
sns.set_style('dark')                          # set the visual style, gery background for better visibility
plt.figure(figsize=(10, 4))                    # plot window, set width = 10, height = 4

sns.scatterplot(x='Age', y='Pclass', data=titanic_df, alpha=0.6, s=80)  # create a scatter plot of age vs. passenger class, set transparency and marker
plt.gca().invert_yaxis()                                                # invert y-axis to show 1st class at the top

# add labels and title
plt.title('Passenger Age vs. Class', fontsize=14)
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Passenger Class', fontsize=12)
plt.yticks([1, 2, 3], ['1st Class', '2nd Class', '3rd Class'])         # set Class labels at exactly these three points on the vertical axis."

plt.tight_layout()                                                     # automatically adjust the spacing, to fit perfectly within the figure area
plt.show()                                                             # show the plot
```

Observations:
- 1st class passengers tend to have a wider age range.
- 3rd class appears to have more younger passengers and fewer elderly individuals.
